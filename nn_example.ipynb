```python
import numpy as np
import matplotlib.pyplot as plt
import modules.nn as nn
import modules.loaddata
```

```python
# This should be generated rather than manual
param_configurations = [
    {
        "depth": 3,
        "epochs": 100
    },
    {
        "depth": 5,
        "epochs": 100
    }
    #etc...
]

ab_initio_q, ab_initio_vals = modules.loaddata.load_ab_initio()

validate_qvals = [150, 250, 350]

validate_indices = np.where(np.isin(ab_initio_q, validate_qvals))
validate_vals = ab_initio_vals[validate_indices]
validate_q = ab_initio_q[validate_indices]

training_indices = np.where(not np.isin(ab_initio_q, validate_qvals))
train_vals = ab_initio_vals[training_indices]
train_q = ab_initio_q[training_indices]


for param_configuration in param_configurations:
    nn = nn.NNInterpolator(param_configuration) # Here we can change what interpolator is used. You should implement the methods of the NNInterpolator class in your model
    nn.train((train_q, train_vals), (validate_q, validate_vals))
    
    # See how the prediction looks for validation data
    for (i, q) in enumerate(train_q):
        vals = []
        for w in validate_vals[i][0][0]:
            val_pred = nn.predict(q, validate_vals[i])
            vals.append(val_pred)
        plt.plot(validate_vals[i][0][0], vals, label=f"Predicted for q={q}")
        plt.plot(validate_vals[i][0][0], validate_vals[i][0][1], label=f"True for q={q}")
```
